```python
#ROCK,PAPER,SCISSORS
from tkinter import*
import random
root=Tk()
root.geometry('600x600')
root.resizable(0,0)
root.title('Rock,Paper,Scissors')
root.config(bg ='seashell3')

Label(root, text="Rock,Paper,Scissors game", font="arial 20 bold", bg="seashell2").pack()

#user choice
user_input=StringVar()
Label(root,text="enter total points you want to play for:",font="arial 13",bg="seashell2").place(x=0,y=70)
Label(root,text="enter your choice:Rock or Paper or Scissors",font="arial 15",bg="seashell2").place(x=0,y=110)
tot_points=IntVar()
Entry(root,textvariable=tot_points,width=20,bg="antiquewhite2").place(x=280,y=73)
Entry(root,textvariable=user_input,bg="antiquewhite2").place(x=120,y=150)


#play
result=StringVar()
def play(): 
    tie=0
    win=0
    loss=0
    i=0
    points=tot_points.get()
    for i in range(points):
        #computer's choice
        comp_choice=random.randint(1,3)
        if comp_choice==1:
            comp_choice="rock"
        elif comp_choice==2:
            comp_choice="paper"
        else:
            comp_choice="scissors"
    
        user_input1=user_input.get()
        if comp_choice==user_input1:
            result.set("Tie!Both have choosen same item")
            tie+=1
        elif comp_choice=="rock" and user_input1=="paper":
            result.set("you win,computer have chosen rock")
            win+=1
        elif comp_choice=="rock" and user_input1=="scissors":
            result.set("you loose,computer have chosen rock")
            loss+=1
        elif comp_choice=="paper" and user_input1=="rock":
            result.set("you loose,computer have chosen paper")
            loss+=1
        elif comp_choice=="paper" and user_input1=="scissors":
            result.set("you win,computer have chosen paper")
            win+=1
        elif comp_choice=="scissors" and user_input1=="rock":
            result.set("you win,computer have chosen scissors")
            win+=1
        elif comp_choice=="scissors" and user_input1=="paper":
            result.set("you loose,computer have chosen scissors")
            loss+=1
        else:
            result.set("enter a valid input:rock or paper or scissors")
        i+=1
    
    score="Your score is "+str(win)
    L1.config(text=score)
    if(win>(points//2)):
        tell="Congrats!You win the game"
        L2.config(text=tell)
    elif(win==tie):
        tell="You have tied the game with computer!"
        L2.config(text=tell)
    else:
        tell="You lose the game!"
        L2.config(text=tell)
L1=Label(root).place(x=25,y=300)
L2=Label(root).place(x=25,y=350)
#reset
def reset():
    result.set("")
    user_input.set("")
    L1.set("")
    L2.set("")
#exit
def exit():
    root.destroy()
    
#buttons
Label(root,textvariable=result,font="arial 15").place(x=25,y=250)
Button(root,text="PLAY",command=play,font="arial 10",bg="seashell2").place(x=150,y=190)
Button(root,text="RESET",command=reset,font="arial 10",bg="seashell2").place(x=70,y=400)
Button(root,text="EXIT",command=exit,font="arial 10",bg="seashell2").place(x=230,y=400)  

root.mainloop()
```
